```python
from bs4 import BeautifulSoup as bs
```

```python
from requests import get
url = 'http://viraindo.com/lcd.html'
response = get(url)
```

```python
text =response.text
```

```python
soup = bs(text, 'html.parser')

#print(soup.prettify())
```

```python
tds=soup.find_all('td')
```

```python
i=0
string=[]
for i in range(0,len(tds),2):
    if tds[i].text.strip('\n').strip()!='' and tds[i+1].text.strip('\n').strip()!='' and (tds[i+1].text.strip('\n').strip()[0].isdigit()) :
        print(i, tds[i].text.strip('\n').strip(),  tds[i+1].text.strip('\n').strip())
        string+=[tds[i].text.strip('\n').strip()]
    i+=1
```

```python
string
```

```python
for s in string:
    split= ((s.split(' ')))
    print (s)
    if 'ACER' in s:
        print('ACER',split[2])
    if 'ASUS ' in s:
        print('ASUS ',split[1])
```
